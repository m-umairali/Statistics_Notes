## Scatterplot

A scatterplot is a common type of data visualization used in data analysis and statistics. It is a graph that displays the relationship between two **quantitative** variables. Each point on the graph represents the value of the two variables for a single observation in the data set.

In a scatterplot, one variable is plotted along the x-axis and the other variable is plotted along the y-axis. The data points are then plotted at the intersection of the two axes. The resulting graph shows the **pattern or relationship between the two variables.**

A scatterplot can show a variety of different relationships between the variables, such as a **positive or negative** correlation, a linear or nonlinear relationship, or no relationship at all. For example, a scatterplot might show a positive correlation between the number of hours spent studying and the grade earned on a test. This means that as the number of hours spent studying increases, the grade earned on the test also tends to increase.

Scatterplots can be useful for several reasons. First, they can help to identify **patterns or trends** in the data that might not be apparent from just looking at a table of numbers. Second, they can help to identify **outliers** or unusual data points that might be worth investigating further. Finally, they can be used to check assumptions about the data and the statistical models used to analyze it.

When creating a scatterplot, it's important to choose appropriate scales for the x-axis and y-axis, label the axes clearly, and add a title to the graph. Additionally, it can be useful to add a regression line or other markers to help highlight any patterns or trends in the data.

Overall, scatterplots are a versatile and powerful tool for exploring relationships between variables in a data set.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the Titanic dataset
titanic = sns.load_dataset("titanic")

# create a scatterplot using seaborn
sns.scatterplot(x="age", y="fare", data=titanic, hue="class", size="class",
                sizes=(15, 25), alpha=0.9, palette="Dark2")

# set the plot title and axis labels
plt.title("Age vs Fare on the Titanic by Class")
plt.xlabel("Age")
plt.ylabel("Fare")

# add legend
plt.legend(title="Class", loc="upper right")

# display the plot
plt.show()
```

## Covariance 



Covariance is a statistical concept that only tells the direction not the relationship between two variables. In particular, it measures how much two variables change together. <br>Covariance is calculated by taking the average of the product of the deviations of two variables from their respective means. 

1. Positive covariance occurs when the two variables increase or decrease together. This means that as one variable increases, so does the other, and vice versa.

2. Negative covariance occurs when one variable increases as the other decreases. This means that as one variable increases, the other decreases, and vice versa.

3. No covariance occurs when there is no relationship between the two variables. This means that the variables are independent of each other, and the value of one variable does not affect the value of the other variable.

Mathematical formula of deriving the covariance is given below:

$$\operatorname{cov}(X,Y) = \frac{\sum\limits_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{n - 1}$$

Let's take a look at an example using the seaborn library in Python. We'll use the famous iris dataset, which contains measurements of different parts of the iris flower.

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Calculate the covariance matrix for the iris dataset
cov_matrix = iris.cov()

# Positive covariance plot
sns.set(style="ticks")
sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="reg")
sns.despine()
plt.subplots_adjust(top=0.95)
plt.suptitle(f"Positive covariance: {cov_matrix.loc['sepal_length', 'petal_length']:.3f}", fontsize=12)

# Negative covariance plot
sns.jointplot(x="sepal_width", y="petal_length", data=iris, kind="reg")
sns.despine()
plt.subplots_adjust(top=0.95)
plt.suptitle(f"Negative covariance: {cov_matrix.loc['sepal_width', 'petal_length']:.3f}", fontsize=12)

# No covariance plot
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="reg")
sns.despine()
plt.subplots_adjust(top=0.95)
plt.suptitle(f"No covariance: {cov_matrix.loc['sepal_length', 'sepal_width']:.3f}", fontsize=12)

plt.show()
```

## Disadvantage:
Covariance only measures the strength of a linear relationship between two variables, and it does not provide any information about the causal relationship between them. It doesnt tell is the relation between two variables strong or weak.

## Correlation
Correlation is the statistical relationship between two or more variables. It measure the degree to which two variables are related and how they tend to change togather. More simply it quantiy the strength and weakness of the relationship

Correlation is usually measured by the correlation coefficient (Pearson's correlation coefficien), which is a value between -1 and 1. A positive correlation means that the variables move in the same direction, while a negative correlation means they move in opposite directions. A correlation of 0 means that there is no linear relationship between the variables.

The formula for correlation coefficient (r) is:

$$ r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}} $$

where $x$ and $y$ are two variables, $n$ is the sample size, $\bar{x}$ and $\bar{y}$ are the means of $x$ and $y$ respectively.

## Correlation vs Causation
Correlation and causation are two concepts that are often used interchangeably, but they have distinct meanings in statistics and scientific research. Understanding the difference between correlation and causation is important to make accurate conclusions based on data. In this blog post, we will explore the definitions of correlation and causation and provide real-world examples to help illustrate the concepts.


Correlation:

Correlation is a statistical measure that describes the strength and direction of the relationship between two variables. It measures the degree to which two variables are related to each other, but it does not imply that one variable causes the other. Correlation is often measured using the correlation coefficient, which ranges from -1 to 1. A correlation coefficient of 1 indicates a perfect positive correlation, a correlation coefficient of -1 indicates a perfect negative correlation, and a correlation coefficient of 0 indicates no correlation between the variables.
For example, let's say we want to examine the correlation between the amount of rainfall and the yield of crops in a particular region. We collect data on the amount of rainfall and the crop yield for several years and calculate the correlation coefficient. If we find that there is a strong positive correlation between rainfall and crop yield, this means that when there is more rainfall, the crop yield tends to be higher. However, this does not necessarily mean that rainfall causes an increase in crop yield. There may be other factors that influence crop yield, such as soil quality, temperature, or pests.

Causation:

Causation refers to a relationship between two variables where one variable causes the other to change. In order to establish causation, we need to demonstrate that changes in one variable are directly responsible for changes in the other variable. This can be challenging to prove, especially in complex systems where many variables interact with each other.

For example, let's say we want to examine the causal relationship between smoking and lung cancer. Many studies have found a strong correlation between smoking and lung cancer, but this does not necessarily mean that smoking causes lung cancer. It is possible that other factors, such as genetics or exposure to environmental toxins, may also play a role in the development of lung cancer. In order to establish causation, we would need to conduct carefully controlled experiments that isolate the effects of smoking on lung cancer.

Real-World Examples:

Let's explore some real-world examples to illustrate the difference between correlation and causation.
Ice Cream and Crime Rates: Studies have found a positive correlation between ice cream sales and crime rates. This may sound counterintuitive, but it is important to note that correlation does not imply causation. It is unlikely that ice cream sales directly cause crime rates to increase. Instead, both variables may be influenced by a third factor, such as temperature. In hot weather, people may be more likely to buy ice cream and more likely to engage in criminal activity.

Education and Income: There is a strong positive correlation between education and income. This means that people with higher levels of education tend to earn more money. However, this does not necessarily mean that education causes people to earn more money. It is possible that other factors, such as natural aptitude or access to resources, may also influence income.

Seat Belt Use and Car Accidents: Studies have found a negative correlation between seat belt use and car accidents. This means that people who wear seat belts are less likely to be involved in car accidents. In this case, the correlation suggests a causal relationship, as wearing a seat belt can help protect people from injury in the event of an accident.

It is important to note that correlation does not imply causation. Just because two variables are correlated does not mean that one variable causes the other. There may be other factors involved that are causing the observed correlation. In order to establish causation, we need to conduct experiments or use other methods to isolate the effect of one variable on another.
