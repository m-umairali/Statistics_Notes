## Graphs for Bivariat Analysis
Bivariat analysis is between 
- Categorical and Categorical columns
- Numerical and numerical columns
- Numerical and categorical columns

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
```

## Categorical - Categorical 
- Contigency table/crosstab
    - A contingency table is a type of table used in statistics to summarize the relationship between two categorical variables. It shows the frequency distribution of the categories of one variable against the categories of the other variable. Contingency tables are commonly used to investigate the relationship between two categorical variables, to test for independence or association between the two variables, and to identify patterns or trends in the data.

```python
survived_crosstab  = pd.crosstab(df['survived'], df['pclass'])
survived_crosstab
```

```python
sns.set_style('whitegrid')
survived_crosstab.plot(kind='bar', stacked=True, color=['#F5B7B1', '#ABEBC6', '#AED6F1'])

# Set the plot title and labels
plt.title('Survival by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```

This table tells us how many people of certain class were died or alived

## Numerical - Numerical
For like always we use scatter plot

```python
plt.scatter(df['age'], df['fare'])

# Set the plot title and labels
plt.title('Scatter Plot of Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show the plot
plt.show()
```
