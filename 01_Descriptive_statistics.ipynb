## What is Statistics
**Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data. It involves methods and techniques for summarizing and describing data, making inferences and predictions from data, and testing hypotheses and making decisions based on data.**

Statistics can be applied in various fields such as business, economics, medicine, psychology, social sciences, and many others. Some common statistical concepts include probability theory, hypothesis testing, regression analysis, and data visualization. The use of statistical methods can help to uncover patterns and relationships in data, as well as provide insights and solutions to problems in various fields.

Examples:

- Business: A company may use statistical methods to analyze sales data and identify trends in customer behavior, in order to make decisions about marketing strategies and product development.

- Medicine: Medical researchers may use statistical methods to analyze clinical trial data and determine whether a new treatment is effective in curing a disease or improving patient outcomes.

- Psychology: Psychologists may use statistical methods to analyze data from experiments and studies, in order to draw conclusions about human behavior and cognitive processes.

- Social Sciences: Sociologists may use statistical methods to analyze survey data and identify patterns of social behavior, such as voting patterns or attitudes towards certain social issues.

- Economics: Economists may use statistical methods to analyze economic data and make forecasts about future economic trends, such as inflation rates or stock market performance.

- Sports: Sports analysts may use statistical methods to analyze player performance data and make predictions about team outcomes, such as which team is likely to win a championship.

- Environmental Science: Environmental scientists may use statistical methods to analyze data on pollution levels and climate change, in order to identify trends and develop solutions to environmental problems.

## Types of Statistics
1. Descriptive:  It refers to the methods used to summarize and describe data, in order to gain a better understanding of its characteristics. This includes measures such as mean, median, mode, range, standard deviation, and variance, as well as data visualization techniques such as histograms, scatter plots, and box plots. Descriptive statistics can help to identify patterns and trends in data, as well as outliers or anomalies.

2. Inferential: It refers to the methods used to make inferences and draw conclusions about a population based on a sample of data. This includes techniques such as hypothesis testing, confidence intervals, and regression analysis. Inferential statistics can help to answer questions such as whether a particular treatment is effective in curing a disease, or whether there is a significant difference between the means of two groups.
## Inferential Statistics:
Inferential statistics is a branch of statistics that involves using sample data to make inferences or draw conclusions about a larger population. It is a powerful tool that enables researchers to make predictions, test hypotheses, and estimate population parameters based on sample data. Some of the key topics in inferential statistics include:

Hypothesis testing: Hypothesis testing is a statistical method for determining whether there is a significant difference between two or more groups, or whether a particular effect is real or due to chance. It involves setting up a null hypothesis (a statement that assumes no effect or difference) and an alternative hypothesis (a statement that assumes there is an effect or difference), and using statistical tests to determine which hypothesis is supported by the data.

Confidence intervals: Confidence intervals are a range of values around a sample statistic (such as the mean or proportion) that are likely to contain the true population parameter with a certain degree of confidence. They are used to estimate population parameters and determine the precision of sample estimates.

Regression analysis: Regression analysis is a statistical method for modeling the relationship between two or more variables. It involves fitting a regression equation to the data and using the equation to predict the value of one variable based on the values of the other variables.

Analysis of variance (ANOVA): ANOVA is a statistical method for comparing the means of two or more groups. It tests whether there is a significant difference between the means of the groups, and helps to identify which group(s) differ from the others.

Bayesian statistics: Bayesian statistics is an approach to inferential statistics that uses probability theory to make inferences about population parameters. It involves updating prior knowledge with new data to arrive at a posterior probability distribution, which represents the updated belief about the parameter.

## Population VS Sample
![](https://www.voxco.com/wp-content/uploads/2021/04/Population-vs-Sample-cvr.jpg)

In statistics, a population is the entire group of individuals, objects, or events that we are interested in studying. A sample, on the other hand, is a subset of the population that is selected for analysis. The sample is used to make inferences about the population, as it is often impractical or impossible to study the entire population.

For example, the population of Pakistan is approximately 225 million people. It would be impossible to collect data on every single person in Pakistan, so we might choose to take a sample of the population instead.

Suppose we are interested in studying the average income of people living in Pakistan. We could take a random sample of 1000 individuals from the population of Pakistan, and collect information on their income levels. This sample of 1000 individuals would be used to make inferences about the income of the entire population of Pakistan.

In this example, the population is the entire group of people living in Pakistan, while the sample is the subset of 1000 individuals that were selected for analysis. By studying this sample, we can draw conclusions about the average income of the population of Pakistan as a whole. The goal of statistical analysis is to ensure that the sample is representative of the population, so that the inferences we make are accurate and reliable.

## Things to be considered while Sampling
When conducting a statistical study, it is important to be careful about the sampling process to ensure that the sample is representative of the population of interest. Here are some things to be careful about while sampling:

Selection bias: Selection bias occurs when the sample is not representative of the population due to the way the sample was selected. For example, if a study only selects participants from a specific geographic region, the results may not be applicable to the population as a whole.

Non-response bias: Non-response bias occurs when individuals who are selected for the sample do not respond or participate in the study. This can result in a biased sample if those who do not respond are different from those who do respond in important ways.

Sampling error: Sampling error refers to the natural variability that occurs between different samples of the same size. It is important to take into account the potential for sampling error and use statistical methods to estimate the precision of the sample estimates.

Sample size: The size of the sample is an important consideration when conducting a statistical study. A sample that is too small may not be representative of the population, while a sample that is too large may be impractical or unnecessary.

Random sampling: To ensure that the sample is representative of the population, it is important to use random sampling methods. Random sampling helps to minimize bias and ensure that each member of the population has an equal chance of being selected for the sample.

## Mean
The mean is a measure of central tendency that represents the average of a dataset. It is calculated by adding up all the values in the dataset and dividing the sum by the total number of values

![](https://www.onlinemathlearning.com/image-files/population-mean.png)


The mean may not be a suitable measure of central tendency in cases where the dataset has extreme values, or outliers, that are significantly different from the rest of the values in the dataset. In such cases, the mean can be influenced heavily by these outliers and may not be representative of the typical value in the dataset.

- For example, consider the following dataset:
- ``10, 20, 30, 40, 50, 100``
- The mean of this dataset can be calculated as follows:
- `mean = (10 + 20 + 30 + 40 + 50 + 100) / 6 = 33.33`

In this case, the value of 100 is an outlier that is much larger than the other values in the dataset. As a result, the mean is higher than what most of the values in the dataset would suggest, and it is not a good measure of central tendency. In this case, the median, which is the middle value in the dataset, might be a better measure of central tendency because it is not affected by outliers. The median for this dataset would be 35, which is a more representative value of the typical value in the dataset.

Therefore, it is important to consider the presence of outliers when deciding which measure of central tendency to use, and the mean may not be the best choice in datasets with extreme values. So you have to be very careful while descriptive statistics that you have outliers or not.


## Median
The median is a measure of central tendency that represents the middle value of a dataset when the values are arranged in order. In other words, it is the value that separates the dataset into two equal halves, with half of the values above the median and half below it.

To calculate the median of a dataset, the values are first arranged in order from smallest to largest or largest to smallest. If the dataset has an odd number of values, then the median is simply the middle value in the ordered dataset. For example, consider the following dataset:

`10, 20, 30, 40, 50`
- In this case, the median is 30, which is the middle value of the dataset.

If the dataset has an even number of values, then the median is the average of the two middle values in the ordered dataset. For example, consider the following dataset:

`10, 20, 30, 40, 50, 60`
- In this case, the middle values are 30 and 40, so the median is the average of these two values:
`median = (30 + 40) / 2 = 35`

The median is a useful measure of central tendency when the dataset has extreme values, or outliers, because it is not influenced by these values. The median is a better measure of central tendency than the mean in such cases because it gives a more representative value of the typical value in the dataset.

## Mode
The mode is a measure of central tendency that represents the value that occurs most frequently in a dataset. In other words, it is the value that has the highest frequency in the dataset.

For example, consider the following dataset:

`10, 20, 30, 30, 40, 40, 40, 50`

In this case, the mode is 40, which occurs three times in the dataset, more than any other value.

If a dataset has no value that occurs more than once, then it has no mode. If there are multiple values that occur with the same highest frequency, then the dataset is said to have multiple modes.

The mode is a useful measure of central tendency for datasets with categorical or discrete data, such as colors, shapes, or the number of siblings a person has. It can also be useful for datasets with continuous data that are grouped into categories, such as age groups or income brackets.

However, the mode is less commonly used than the mean and median as a measure of central tendency in statistical analysis, and it may not always give a good representation of the typical value in a dataset. This is because the mode only considers the value that occurs most frequently and ignores the other values in the dataset.

## Weighted Mean
The weighted mean is a type of mean that takes into account the different weights or importance of the values in a dataset. It is calculated by multiplying each value in the dataset by its corresponding weight, summing up these products, and dividing the sum by the total weight of the dataset.

Mathematically, the weighted mean can be expressed as:

`weighted mean = (w1*x1 + w2*x2 + ... + wn*xn) / (w1 + w2 + ... + wn)`

where x1, x2, ..., xn are the values in the dataset and w1, w2, ..., wn are the corresponding weights.

For example, consider a dataset of test scores, where the scores range from 0 to 100 and each score has a corresponding weight that represents the percentage of the overall grade that the test is worth. The dataset and corresponding weights are:

`Test score: 70, 80, 90, 95`

`Weight (% of grade): 20%, 30%, 40%, 10%`

To calculate the weighted mean of these scores, we first multiply each score by its corresponding weight:

`70*0.2 = 14`

`80*0.3 = 24`

`90*0.4 = 36`

`95*0.1 = 9.5`

Then, we sum up these products:

`14 + 24 + 36 + 9.5 = 83.5`

Finally, we divide the sum by the total weight of the dataset:

`0.2 + 0.3 + 0.4 + 0.1 = 1`

`weighted mean = 83.5 / 1 = 83.5`

So the weighted mean of the test scores is `83.5`

The weighted mean is a useful measure of central tendency when the values in a dataset have different levels of importance or significance. It is commonly used in areas such as finance, economics, and education to calculate averages that take into account the different weights or importance of the values in a dataset.

## Measure of Dispersion
A measure of dispersion, also known as a measure of variability or spread, is a statistical metric that describes the extent to which the values in a dataset vary from each other. It quantifies how much the data points deviate from the central tendency, which is usually represented by a measure of location such as the mean or median.

Some commonly used measures of dispersion include:

- Range: The range is the simplest measure of dispersion and is defined as the difference between the highest and lowest values in the dataset. It provides a quick and easy way to understand the spread of the data, but is highly sensitive to outliers.

- Variance: The variance is a more robust measure of dispersion that calculates the average of the squared differences between each data point and the mean of the dataset. It is calculated as the sum of the squared deviations from the mean divided by the number of observations minus 1.

- Standard deviation: The standard deviation is the square root of the variance and is often used instead of the variance because it is expressed in the same units as the data. It measures the average deviation of the data points from the mean and provides a more intuitive measure of the spread of the data.

- The coefficient of variation (CV) is a measure of relative dispersion that is often used to compare the variability of two or more datasets that have different units or scales.

- Interquartile range (IQR): The IQR is a measure of dispersion that is based on the quartiles of the dataset. It is calculated as the difference between the third quartile (the 75th percentile) and the first quartile (the 25th percentile) of the dataset. The IQR is less sensitive to outliers than the range and provides a better measure of the spread of the middle 50% of the data.

- Mean absolute deviation (MAD): The MAD is a measure of dispersion that calculates the average of the absolute differences between each data point and the mean of the dataset. It is less sensitive to outliers than the variance and provides a better measure of the spread of the data when the dataset contains extreme values.

These measures of dispersion are useful for describing the spread of the data and for comparing the variability of different datasets. Depending on the specific characteristics of the dataset and the research question, different measures of dispersion may be more appropriate to use.

The range can be a useful statistic to quickly assess the variability in a dataset, but it is sensitive to outliers and may not provide a complete picture of the spread of the data

## Variance

variance is a measure of how spread out a set of data is. Specifically, it measures the average of the squared differences from the mean.

To calculate the variance of a dataset, you follow these steps:

1. Find the mean (average) of the data set.
3. For each data point, subtract the mean and square the result.
3. Calculate the average of the squared differences.

Suppose we have the following dataset representing the number of hours each employee worked in a week:

`{38, 40, 42, 35, 37}`

To calculate the variance, we first need to find the mean of the dataset:

`Mean = (38+40+42+35+37) / 5 = 38.4`

Next, we need to find the squared differences from the mean for each data point:

`(38 - 38.4)^2 = 0.16`

`(40 - 38.4)^2 = 2.56`

`(42 - 38.4)^2 = 12.96`

`(35 - 38.4)^2 = 11.56`

`(37 - 38.4)^2 = 2.56`

Now we calculate the average of these squared differences:

`Variance = (0.16 + 2.56 + 12.96 + 11.56 + 2.56) / 5 = 5.36`

So the variance of this dataset is `5.36`. This tells us that the data points are somewhat spread out from the mean of `38.4`. However, without knowing the units of measurement (e.g., hours, dollars, etc.), it's hard to interpret the magnitude of the variance.

## Standard Deviation
Standard Deviation is the square root of the variance. It is a widely used measure of dispersion that is useful in describing the shape of the distribution. 

**Note:**

Standard deviation and variance are both measures of the spread or dispersion of a dataset, but they differ in how they are calculated and the way they express the dispersion.

Variance is the average of the squared differences from the mean, while the standard deviation is the square root of the variance. Both measures give you an idea of how much the data points deviate from the mean, but the standard deviation is more commonly used because it has the same unit of measurement as the original data, making it easier to interpret.

For example, if you have a dataset representing the weights of a sample of apples in kilograms, the variance of the dataset will be in units of kilograms squared, which is not easily interpretable. However, if you take the square root of the variance to get the standard deviation, the result will be in units of kilograms, which is more meaningful and easier to understand.
