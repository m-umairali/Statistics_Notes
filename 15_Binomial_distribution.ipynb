## Binomial Distributions
The binomial distribution is a discrete probability distribution that describes the probability of a fixed number of successes in a fixed number of independent Bernoulli trials. A Bernoulli trial is a random experiment that can have two possible outcomes: success or failure. In a binomial distribution, each trial has the same probability of success, denoted by p.

The binomial distribution is characterized by two parameters: n, which represents the number of trials, and p, which represents the probability of success in each trial. The probability mass function of the binomial distribution is defined as:

$$P(X = k) = {n \choose k} \cdot p^k \cdot (1-p)^{n-k}$$


where:

$n$ is the number of independent trials

$k$ is the number of successful outcomes

$p$ is the probability of success in a single trial

$1-p$ is the probability of failure in a single trial

${n\choose k}$ is the binomial coefficient, which is equal to $\frac{n!}{k!(n-k)!}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Set the parameters for the binomial distribution
n = 10    # number of trials
p = 0.5   # probability of success

# Generate the binomial distribution
x = np.arange(0, n+1)
binom_dist = binom.pmf(x, n, p)

# Visualize the binomial distribution
plt.bar(x, binom_dist)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial distribution')
plt.show()
```

This code sets the number of trials `n` and the probability of success `p`, generates the probability mass function of the binomial distribution using the `binom.pmf()` function from the `scipy.stats` module.

## Criteria
1. The experiment consists of a fixed number of trials, denoted by n.
2. Each trial is independent, meaning that the outcome of one trial does not affect the outcome of any other trial.
3. There are only two possible outcomes for each trial, commonly denoted as success or failure.
4. The probability of success for each trial is constant and denoted by p.

## Real life examples:
- Tossing a coin: The outcome of a coin toss can be either heads or tails, which makes it a binary event. If you toss a coin n times, the number of times it lands heads can be modeled using a binomial distribution.

- Quality control: In manufacturing, the quality of products is often checked by taking a sample and checking if it meets certain criteria. If the sample size is large and the proportion of defective items in the population is known, the number of defective items in the sample can be modeled using a binomial distribution.

- Election results: During an election, voters either choose a particular candidate or they don't. If the total number of voters is large and the proportion of voters who support a particular candidate is known, the number of votes that the candidate receives can be modeled using a binomial distribution.

- Medical trials: In a medical trial, a group of participants either responds to a treatment or they don't. If the total number of participants is large and the proportion of participants who respond to the treatment is known, the number of participants who respond to the treatment can be modeled using a binomial distribution.
