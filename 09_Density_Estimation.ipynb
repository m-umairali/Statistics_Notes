## Density Estimation
Density estimation is the process of estimating the probability density function of a random variable from a given set of data.
The goal of density estimation is to estimate the underlying probability density function of a random variable from a given set of observations, and to use this estimate to make inferences about the population from which the observations were drawn. <br>
There are many methods for estimating the probability density function of a random variable, including kernel density estimation, parametric density estimation, and non-parametric density estimation. 

## Parametric Density Estimation:
Parametric estimation is a method of estimating the probability distribution of a dataset using a specified mathematical model with a fixed set of parameters. This method assumes that the data follows a particular probability distribution, and the parameters of that distribution are estimated from the data. These parameters are then used to define the probability distribution of the data. 
<br>This approach is often used when there is prior knowledge about the underlying distribution (normal, exponential, poisson etc) of the data or when the data follows a known theoretical distribution. The advantage of parametric estimation is that it can provide a more precise and accurate estimation of the probability distribution of the data if the assumed model is accurate.



## Example: 
Let's take an example to understand parametric density estimation better. Suppose we have a dataset of the weights of apples in a fruit market. We want to estimate the probability density function (PDF) of the weight of apples in the market.

Now, if we assume that the weight of apples follows a normal distribution, then we can use parametric density estimation to estimate the PDF of the weights.

We first estimate the mean and standard deviation of the weight of apples from our dataset. Let's assume that the mean weight of apples is 150 grams and the standard deviation is 20 grams.

Then we use the probability density function of the normal distribution to estimate the PDF of the weights. The PDF of a normal distribution is given by:

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

where:

x is the weight of the apple <br>
μ is the mean weight of apples in the market (150 grams in our example) <br>
σ is the standard deviation of the weight of apples in the market (20 grams in our example) <br>
We can use the above formula to estimate the PDF of the weight of apples in the market. We can then plot the estimated PDF using a seaborn graph to visualize the distribution of the weights.

Here's the Python code to implement parametric density estimation for our example:

```python
import numpy as np
import seaborn as sns

# Load the dataset of weights of apples
weights = np.array([140, 150, 160, 170, 180, 190, 200])

# Estimate the mean and standard deviation of the weights
mean = np.mean(weights)
std_dev = np.std(weights)

# Define the probability density function of the normal distribution
def normal_pdf(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2*np.pi))) * np.exp(-((x - mean)**2) / (2 * std_dev**2))

# Generate a range of weights to plot the PDF
x_range = np.linspace(100, 250, 100)

# Calculate the PDF of the weights using the normal distribution
pdf_values = [normal_pdf(x, mean, std_dev) for x in x_range]

# Plot the estimated PDF using seaborn
sns.lineplot(x=x_range, y=pdf_values)
```

The above code will generate a plot of the estimated PDF of the weight of apples in the market assuming a normal distribution. We can see the shape of the distribution and the most probable weight range for the apples.

Parametric density estimation is useful when we have prior knowledge about the underlying distribution of the data. It can help us estimate the PDF of the data accurately and efficiently using mathematical formulas.

## Non Parametric Density Estimation (KDE)
Sometimes, the distribution is not clear and it is not one of the famous distribution.

Non-parametric density estimation is a method of estimating the probability density function of a random variable without assuming any specific parametric form for the density function. Instead, non-parametric methods rely on estimating the density function directly from the observed data.

One common non-parametric method for density estimation is kernel density estimation (KDE). KDE works by placing a kernel (e.g., a Gaussian distribution) on each observed data point, and then summing these kernels to obtain an estimate of the density function.

### Advantages:

Non-parametric density estimation is a flexible method that can work well for a wide range of data distributions without assuming any specific underlying model.
It is easy to use and does not require any knowledge about the underlying distribution.
Non-parametric density estimation can handle multimodal distributions, which is not possible with some parametric methods.
It is robust to outliers.

### Disadvantages:

Non-parametric density estimation can be computationally expensive, especially for large datasets, and may require more memory than parametric methods.
The quality of the estimation depends on the choice of the kernel function and the bandwidth parameter, which can be subjective and may require some tuning.
Non-parametric density estimation may not be as accurate as parametric methods when the underlying distribution is well-defined and known.
Non-parametric density estimation can be sensitive to the choice of sample size, and smaller sample sizes may lead to poor estimates.

## Kernel Density Estimation
Kernel Density Estimation (KDE) is a non-parametric method used for estimating the probability density function (PDF) of a random variable. It is based on the idea of placing a kernel function at each data point and then adding up the kernel functions to obtain an estimate of the PDF.
