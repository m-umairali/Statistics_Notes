## Graphs For Univariat Analysis
Ploting graphs of univariat is depend upon the type of data whether it is categorical or numerical.

## For categorical

If you have categorical column you can make 
1. A frequency distribution table is a table that shows the number of occurrences of each value or category in a dataset. It typically includes two columns: one for the values or categories and another for the frequency, which is the number of times each value or category occurs in the dataset. Frequency distribution tables can help to summarize the distribution of a dataset and identify any patterns or outliers.

2. Relative Frequency:
Relative frequency is the proportion or percentage of observations in a dataset that fall within a particular category or value. It is calculated by dividing the frequency of the category or value by the total number of observations in the dataset. Relative frequency is useful for comparing the frequencies of different categories or values in a dataset and can help to identify the most common or rare categories.

3. Cumulative Frequency:
Cumulative frequency is the total frequency of all values or categories up to a certain point in the dataset. It is calculated by summing the frequencies of all values or categories up to a particular value or category. Cumulative frequency is useful for analyzing the distribution of a dataset and can help to identify the percentage of observations that fall within certain ranges of values or categories.

## Frequency Distribution Table


```python
import pandas as pd

# Create the dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Gina', 'Hannah', 'Isaac', 'Jack'],
    'Destination': ['Beach', 'City', 'Mountain', 'Beach', 'City', 'Mountain', 'Beach', 'City', 'Mountain', 'Beach'],
    'Budget': ['Low', 'High', 'Medium', 'Low', 'Low', 'Medium', 'Medium', 'High', 'Low', 'High'],
    'Activity': ['Relaxation', 'Sightseeing', 'Hiking', 'Relaxation', 'Sightseeing', 'Hiking', 'Relaxation', 'Sightseeing', 'Hiking', 'Sightseeing']
}

# Create a pandas dataframe
df = pd.DataFrame(data)

# Print the dataframe
print(df)

import seaborn as sns

# Create a bar chart of the frequency distribution
sns.countplot(x='Destination', data=df)
```

## Reletive Frequency

```python
destination_freq = df['Destination'].value_counts(normalize=True)

# Create a pie chart of the relative frequency
plt.pie(destination_freq, labels=destination_freq.index, autopct='%1.1f%%')
plt.title('Relative Frequency of Destination')
plt.axis('equal')
plt.show()
```

## Cumilative Frequency

```python
# Calculate the cumulative frequency of each destination
destination_freq = df['Destination'].value_counts(normalize=True)
cumulative_freq = destination_freq.cumsum()

# Create a line chart of the cumulative frequency
plt.plot(cumulative_freq.index, cumulative_freq.values)
plt.title('Cumulative Frequency of Destination')
plt.xlabel('Destination')
plt.ylabel('Cumulative Frequency')
plt.show()
```

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
# Create a histogram of the 'Age' column using Seaborn
sns.histplot(data=titanic, x='age')

# Add axis labels and a title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Passengers by Age')

# Show the plot
plt.show()
```
