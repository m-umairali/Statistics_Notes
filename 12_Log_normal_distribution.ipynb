## Log Normal Distribution:
Log-normal distribution is a continuous probability distribution of a random variable <mark>whose logarithm is normally distributed.</mark> This means that if X is a log-normally distributed random variable, then Y = log(X) has a normal distribution. The distribution is skewed to the right, meaning that it has a long right tail.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a right-skewed dataset
np.random.seed(1)
data = np.random.exponential(scale=10, size=1000)

# Plot the histogram
plt.hist(data, bins=50, density=True)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

```

```python
# Take the natural logarithm of the data
log_data = np.log(data)

# Plot the histogram
plt.hist(log_data, bins=50, density=True)
plt.xlabel('Log values')
plt.ylabel('Frequency')
plt.show()

```

by taking the logarithm of a right-skewed dataset, we can transform it into a roughly normal distribution

#### Note: All the right skewed are qualified for normal distributions if their graph, after taking the log, is equal to normal distribtuions. And also note that all rightly skewed are not always normal distributions.

## Real life examples:
1. The length of comments posted in Internet discussion forums follows a log-normal distribution
2. Measures of size of living tissue (length, skin area, weight)
3. In economics, there is evidence that the income of 97%–99% of the population is distributed log-normally
4. In reliability analysis, the log-normal distribution is often used to model times to repair a maintainable system.
