## Log Normal Distribution:
Log-normal distribution is a continuous probability distribution of a random variable <mark>whose logarithm is normally distributed.</mark> This means that if X is a log-normally distributed random variable, then Y = log(X) has a normal distribution. The distribution is skewed to the right, meaning that it has a long right tail.

<img src="https://docs.analytica.com/images/d/d9/LogNormal%28median%3D3%2Cstddev%3D2%29.png" alt="drawing" width="45%" height="400"/>



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a right-skewed dataset
np.random.seed(1)
data = np.random.exponential(scale=10, size=1000)

# Plot the histogram
plt.hist(data, bins=50, density=True)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

```

```python
# Take the natural logarithm of the data
log_data = np.log(data)

# Plot the histogram
plt.hist(log_data, bins=50, density=True)
plt.xlabel('Log values')
plt.ylabel('Frequency')
plt.show()

```

by taking the logarithm of a right-skewed dataset, we can transform it into a roughly normal distribution

#### Note: All the right skewed are qualified for normal distributions if their graph, after taking the log, is equal to normal distribtuions. And also note that all rightly skewed are not always normal distributions.

## Real life examples:
1. The length of comments posted in Internet discussion forums follows a log-normal distribution
2. Measures of size of living tissue (length, skin area, weight)
3. In economics, there is evidence that the income of 97%–99% of the population is distributed log-normally
4. In reliability analysis, the log-normal distribution is often used to model times to repair a maintainable system.

## Pareto distribution
Pareto distribution, named after Italian economist Vilfredo Pareto, is a continuous probability distribution that describes the distribution of wealth, income, or any other quantity where few events account for the majority of the distribution.

The Pareto distribution has two parameters: x_min, the minimum possible value of the quantity being measured, and α, a positive shape parameter that determines the rate of decay of the distribution

Pareto distributions have many real-world applications, such as modeling the distribution of income, wealth, city populations, or the number of website hits. It is important to note that not all phenomena follow a Pareto distribution, and it is crucial to test the fit of the distribution to the data before making any conclusions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Generate a Pareto distribution
x_min = 1
alpha = 2
pareto_dist = pareto(alpha, loc=x_min)

# Plot the probability density function
x = np.linspace(1, 10, 1000)
pdf = pareto_dist.pdf(x)
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()
```

As you can see, the Pareto distribution has a long right tail, and the probability of extreme events (i.e., events far to the right of the mean) is much higher than in a normal distribution.
